## Part 1 - Form Vocabulary
Compute the vocabulary of the training corpus and print it.

````python
# Import  libraries 
import pandas as pd
import numpy as np
import sys
import copy
````

````python
#creating list of vocabulary 
unigrams = []
# open and read corpus file
with open(r"mother-goose-corpus.txt",'r') as f:
    for line in f:  
        for word in line.split():
            unigrams.append(word)
## unigram contains all words in corpus

#print(unigrams)    

# this list is to store all unique words in corpus which we also called vocabulary 
vocab = [] 
#Set function is used to take all the unique elements of the list of unigrams
vocab = list(set(unigrams)) 
print("vocabulary of the training corpus :")
print(vocab)

````

## Part 2 - Compute unigram probability
Calculate the probability of each unigram and print the 2 unigrams with the highest probability.

````python
#lenght of vocabulary 
n=len(unigrams)
#print(n)

#creating list for unigram probability
uniprob = []
for uni in vocab:
    temp = []
    temp.append(uni)
    prob = unigrams.count(uni)/n
    temp.append(prob)
    uniprob.append(temp)
    freq=int(prob*n)
    temp.append(freq)
    

#coverting list to pandas dataframe 
df =pd.DataFrame(uniprob, columns =['Unigram','Probability','freq'])
#printing dataframe of unigrams probabilty 


df.sort_values("Probability", axis = 0, ascending = True, 
                 inplace = True,)

#When inplace = True is used, it performs operation on data and nothing is returned.
#When inplace=False is used, it performs operation on data and returns a new copy of data.
df
````

## Part 3 - MLE Probability Matrix of Bigrams
Construct a probability matrix containing the maximum likelihood estimates (MLEs) of all possible
bigrams and print it out.

````python
# Import text file into object while replacing newline character with space
text_file = open(r"mother-goose-corpus.txt",'r')
corpus = text_file.read()
corpus = corpus.replace('\n' , ' ')
corpus = ' ' + corpus + ' '   # Padding start and end with spaces
````

````python
# To store list of bigram probabilities as 3-tuple
# (first_word, second_word, P(second_word | first_word) )
bigram_prob = []

for first_word in vocab:
  bigram_row = []
  first_word_count = unigrams.count(first_word)

  for second_word in vocab:
    # Construct required bigram to be searched in the corpus.
    # Spaces have to be added in order to avoid problems while searching at the extremes
    # of corpus. 
    req_bigram = " " + first_word + " " + second_word + " "

    # The probability of each bigram is calculated as count(bigram) / count(first word)
    prob = corpus.count(req_bigram) / first_word_count
    bigram_row.append((first_word, second_word, prob))
  bigram_prob.append(bigram_row)

# For pretty-printing the MLE Matrix
mat = []
for row in bigram_prob:
  mat.append([bigram[2] for bigram in row])

print("Maximum Likelihood Estimate matrix:")
display(pd.DataFrame(mat, columns = vocab, index = vocab))
````

## Part 4 - Bigram(s) with highest MLE
What is the most frequent bigram in this corpus?

````python
# List to store all bigram probabilities so that it can be sorted later
# Tuples are: (prob, first word, second word)
bigram_list = []
for row in bigram_prob:
  bigram_list.extend([(bigram_ele[2], bigram_ele[0], bigram_ele[1]) for bigram_ele in row])

# Sort the bigram list
bigram_list.sort()

print("Most frequent bigram(s) is(are):")
counter = len(bigram_list) - 1
while counter >= 0:
  if bigram_list[-1][0] > bigram_list[counter][0]:
    break
  print(bigram_list[counter][1], bigram_list[counter][2], 'with probability', bigram_list[counter][0])
  counter = counter - 1
````

## Part 5 - Laplace smoothing
Construct a probability matrix containing the Laplace smoothed estimates of all possible bigrams
and print it out.

````python
# To contain laplace-smoothed bigram probabilities
# SENTENCE_START = "<s>"
# SENTENCE_END = "</s>"


#removing <s> from vocabulary 
vocab.remove('<s>')
#print(vocab)
smoothed_bigram_prob = []
N = len(vocab)

for first_word in vocab:
  smoothed_bigram_row = []
  first_word_count = unigrams.count(first_word)

  for second_word in vocab:
    req_bigram = " " + first_word + " " + second_word + " "
    smoothed_prob = (corpus.count(req_bigram) + 1) / (first_word_count + N)
    smoothed_bigram_row.append((first_word, second_word, smoothed_prob))
  smoothed_bigram_prob.append(smoothed_bigram_row)

# For pretty-printing the Laplace Smoothed Matrix
mat = []
for row in smoothed_bigram_prob:
  mat.append([bigram[2] for bigram in row])

print("Laplace Smoothed Estimate matrix:")
display(pd.DataFrame(mat, columns = vocab, index = vocab))
````

## Part 7 - Bigram Count of Counts
Construct and show a count-of-counts table for the bigrams.

````python
# For storing count of counts of frequencies of bigrams:
bigram_freq_coc = {}

for first_word in vocab:
  for second_word in vocab:
    req_bigram = " " + first_word + " " + second_word + " "
    count = corpus.count(req_bigram)
    if count in bigram_freq_coc:
      bigram_freq_coc[count] += 1
    else:
      bigram_freq_coc[count] = 1

coc_table = []
for key, value in sorted(bigram_freq_coc.items(), key = lambda p:p[1]):
  coc_table.append([key, value])

print("Count of Counts table")
print(pd.DataFrame(coc_table , columns = ["Counts" , " Number of Counts"]).to_string(index = False))
````
